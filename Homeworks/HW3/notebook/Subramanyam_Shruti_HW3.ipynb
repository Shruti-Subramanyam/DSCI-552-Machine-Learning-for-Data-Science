```python
def dataread(file,lower,upper):
    data = []
    for i in range(lower,upper+1):
        sniffer = csv.Sniffer() 
        seprt = sniffer.sniff(open("../data/AReM/{}/dataset{}.csv".format(file,str(i))).read()).delimiter
        if seprt==',':
            df = pd.read_csv("../data/AReM/{}/dataset{}.csv".format(file,str(i)),sep = ",", skiprows=5,header=None,index_col=None,error_bad_lines=False)
        else:
            df = pd.read_csv("../data/AReM/{}/dataset{}.csv".format(file,str(i)),sep = "\s+", skiprows=5,header=None,index_col=None,error_bad_lines=False)
        df.columns = ['time','avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23']
        data.append(df)
        del df
    cnct = pd.concat(data,axis=0,ignore_index=True)
    return cnct
```

### (c) Feature Extraction

#### i. Research

Time series analysis related features are as follows:<br>
1.Mean - The average of data points in the time series, <br>
2.Median - Its the value that separates the higher half from the lower half of data, <br>
3.Mode - The most frequently occuring value in the time series, <br>
4.Range - The difference between the maximum and the minimum values, <br>
5.Variance - A measure of how spread the datapoints are from the mean, <br>
6.Standard Deviation - The square root of variance, <br>
7.Skweness - A measure of the asymmetry in the data distribution, <br>
8.Kurtosis - A measure of the "tailedness" of the data distribution,<br> 
9.Minimum - The smallest value of the time series, <br>
10.Maximum - The largest value of the time series, <br>
11.1st quartile -  The value below which 25% of the data falls, <br>
12.3rd quartile -  The value below which 75% of the data falls, <br> 
13.Sum - The sum of all data points within the time window, <br>
14.Waveform Length - The cumulative sum of the absolute differences between consecutive data points in the time series. , <br>
15.Zero Crossing - The count of times the time series crosses the zero-axis, indicating how often the signal changes direction  <br>

#### ii. Extraction

```python
flist = []
cnt=0
p=1
for folder in os.listdir("../data/AReM"):
    for k in range(1,len(os.listdir("../data/AReM/{}".format(folder)))+1):
        sniffer = csv.Sniffer()
        seprt = sniffer.sniff(open("../data/AReM/{}/dataset{}.csv".format(folder,str(k))).read()).delimiter
        if seprt==',':
            df = pd.read_csv("../data/AReM/{}/dataset{}.csv".format(folder,str(k)),sep = ",", skiprows=5,header=None,index_col=None,error_bad_lines=False)
        else:
            df = pd.read_csv("../data/AReM/{}/dataset{}.csv".format(folder,str(k)),sep = "\s+", skiprows=5,header=None,index_col=None,error_bad_lines=False)
        df.columns = ['time','avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23']
        df_ft = pd.DataFrame()
        q=1
        for i in df.columns:
            if i == 'time':
                continue
            df_ft.at[cnt,"Instance"] = p
            df_ft.at[cnt,"folder"] = folder
            df_ft.at[cnt,"filename"] = 'dataset'+str(k)
            df_ft.at[cnt,"min{}".format(q)] = df[i].min()
            df_ft.at[cnt,"max{}".format(q)] = df[i].max()
            df_ft.at[cnt,"mean{}".format(q)] = df[i].mean()
            df_ft.at[cnt,"median{}".format(q)] = df[i].median()
            df_ft.at[cnt,"std{}".format(q)] = df[i].std()
            df_ft.at[cnt,"1stquartile{}".format(q)] = np.percentile(df[i],25)
            df_ft.at[cnt,"3rdquartile{}".format(q)] = np.percentile(df[i],75)
            q+=1
        df_ft.Instance = df_ft.Instance.astype("int")
        flist.append(df_ft)
        cnt+=1
        p+=1
        del df_ft

flist = pd.concat(flist,ignore_index=True)
flist
```

#### iii. Standard Deviation

```python
tdfeatures.std()
```

```python
ci=pd.DataFrame()
for column in tdfeatures.columns:
    bsample_size = len(tdfeatures[column])
    bstd = []
    for i in range(1000):
        bootstrap_sample = tdfeatures[column].sample(n = bsample_size, replace = True)
        bstd.append(bootstrap_sample.std())
    bstd.sort()
    left = np.percentile(bstd, (0.1/2)*100)
    right = np.percentile(bstd, 100-(0.1/2)*100)
    ci[column] = [left,right]
    
ci = ci.T
ci.columns = ['LeftTail','RightTail']
ci
```

#### iv. Select Features

From the above mentioned table values of bootstrap and confidence intervals we can day that the three most important time-domain features are:  <br>
* Mean - When there are lots of unusual data points, the mean (average) can still be a useful way to find the typical value in a set of numbers, especially when the numbers are continuous. <br>
* Median - Because there are many unusual data points (outliers), the median is a better way to find the typical value in a set of numbers. This is because outliers can pull the mean towards them, but the median isn't as influenced by them.  <br>
* Standard deviation - Standard deviation is a way to figure out how spread out data is and to spot data points that are far from the average. It also gives us clues about the shape of the data pattern. <br>
as these features helps to unders tand the distribution of data.
