```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# I. Pandas

```python
""" 
    (a) Consider the Salaries.csv file.
    (b) Use the read_csv(...) method from Pandas (Documentation Link) to read data from file Salaries.csv 
        and to copy it into a dataframe.
    (c) Make the column playerID in the csv file as the index column and the first row as the header. 
        Also, skip the second row when reading the file. 
"""

df = pd.read_csv("../data/Salaries.csv",header=[0],skiprows=[1]).set_index('playerID')
```

```python
df.head()
```

```python
"""
    (d) Select the id of the players who are registered in ATL and HOU whose salary is higher than one million.
"""

id = df[((df['teamID'] == 'ATL') | (df['teamID'] == 'HOU')) & (df['salary'] > 1000000)]
id
```

```python
"""
    (e) Use the describe() method to calculate the standard deviation, first quartile, median, 
        third quartile, mean, maximum, and minimum of the salary in team ATL.
"""

df[df["teamID"]=='ATL']['salary'].describe() 
```

```python
"""
    (f) Create a Python dictionary object whose keys are the headers of the dataframe created in the read_csv() exercise 
        and values are Python list objects that contain data corresponding to the headers. 
        (Here, use the iterrows() method to iterate each row of the dataframe and copy it to a dictionary. 
        However, there is an easier way. Learn how the to_dict() method works by yourself later)
"""

dic = {}
for i in df.columns:
    dic[i] = []
for index, row in df.iterrows():
    dic['yearID'].append(row['yearID'])
    dic['teamID'].append(row['teamID'])
    dic['lgID'].append(row['lgID'])
    dic['salary'].append(row['salary'])
print(dic)
```

```python
"""
    (g) Create a dataframe using pd.DataFrameRead (Documentation Link) and from
        the dictionary created in (e). Then, change the header to "a", "b", "c", ... .
"""

dfNew = pd.DataFrame.from_dict(dic)
dfNew.columns =['a','b','c','d']
dfNew
```

# II. Numpy

```python
"""
    (a) Create a 2-dimensional Python list object, then convert it to a Numpy array object.
"""

lst = [549,551,552,553,560,564,566,585]
narray = np.array(lst)
print("Numpy array object is:",narray)
print("Type of numpy array is:",type(narray))
```

```python
"""
    (b) Examine the ndim, shape, size, dtype, itemsize, and data attributes of the numpy
        array object. Make sure you understand their functions.
"""

print("ndim:",narray.ndim)
print("shape:",narray.shape)
print("size:",narray.size)
print("dtype:",narray.dtype)
print("itemsize:",narray.itemsize)
print("flags:",narray.flags)
```

```python
"""
    (c) Learn the dimension concept of an ndarray object by using reshape() and flatten() methods.
"""

rnarray = np.reshape(narray, (4,2))
print("reshaped array:\n",rnarray)
fnarray = rnarray.flatten()
print("\nflatten array:\n",fnarray)
```

```python
"""
    (d) Understand how the slice operation works for 1-D arrays and 2-D arrays and practice by yourself.
"""

print("Slicing for 1-D array [3:6]:\n",narray[3:6])
print("Slicing for 2-D array [0:3, ::]:\n",rnarray[0:3, ::])
```

```python
"""
    (e) Learn operations on ndarray by examining the argmin(), argmax(), min(), max(), mean(), 
        sum(), std(), dot(), square(), sqrt(), abs(). exp(), sign(), and mod() methods. 
        Make yourself comfortable with these methods.
"""

print("argmin:",np.argmin(narray))
print("argmax:",np.argmax(narray))
print("min:",np.min(narray))
print("max:",np.max(narray))
print("mean:",np.mean(narray))
print("sum:",np.sum(narray))
print("std:",np.std(narray))
print("dot:",np.dot(narray,[2,2,2,2,2,2,2,2]))
print("sqaure:",np.square(narray))
print("sqrt:",np.sqrt(narray))
print("abs:",np.abs(narray))
print("exp:",np.exp(narray))
print("sign:",np.sign(narray))
print("mod:",np.mod(narray,7))
```

```python
"""
    (f) Examine the arange(), ones(), zeros(), eye(), linspace(), and concatenate() methods. 
        Make yourself comfortable with these methods.
"""

print("arange:",np.arange(1,45,7))
print("ones:\n",np.ones((3,4)))
print("zeros:\n",np.zeros([4, 3], dtype = int))
print("eye:\n",np.eye(3))
print("linspace:\n",np.linspace(3,5,num=12))
print("concatenate:\n",np.concatenate(([2,3],[4,5])))
```

# III. Matplotlib

```python
"""
    (a) Create two one dimensional arrays x and y and plot y vs x, add title, xlabel, ylabel, grid.
"""

x = np.linspace(-10, 10, num=40)
y = np.array([j ** 2 for j in x])
plt.title("Matplotlib Visualization")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
"""
    (b) Create multiple arrays and plot them with different styles, add legends, add text/mathematical equations on the plot.
"""

x = np.linspace(-10, 10, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])
plt.plot(x, y1, label = "y1=x")
plt.plot(x, y2, label = "y2=x**2")
plt.plot(x, y3, label = "y3=x**3")
plt.title("Multiple style visualization")
plt.text(4.5,320, '$y3 = x^3$', fontsize = 10)
plt.text(6.5,90, '$y2 = x^2$', fontsize = 10)
plt.text(8,-90, '$y1 = x$', fontsize = 10)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.legend()
plt.show()
```

```python
"""
    (c) Create multiple subplots, play around with the figure size, text font/size.
"""

fig, (ax1, ax2, ax3) = plt.subplots(3,figsize=(6,8))
ax1.plot(x, y1)
ax1.set_title("y1=x")
ax2.plot(x, y2)
ax2.set_title("y2=x^2")
ax3.plot(x, y3)
ax3.set_title("y3=x^3")
fig.tight_layout()
plt.show()
```

```python
"""
    (d) Get familiar with get current axis (gca) handle to do the above tasks
"""

ax = plt.gca()
print (ax)
```

```python
"""
    (e) Change the limits on x and y axes, use logarithmic axes to plot
"""

fig, (ax1, ax2,ax3) = plt.subplots(3,figsize=(6,8))
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax1.set_xscale('log')
ax1.set_yscale('log')
ax2.set_xscale('log')
ax2.set_yscale('log')
plt.xlabel('Log X')
plt.ylabel('Log Y')
plt.show()
```
